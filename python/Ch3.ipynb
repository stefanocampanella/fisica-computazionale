```python
import pylab as plt
import numpy as np
```

```python
x = np.linspace(0.,np.pi,500)
plt.plot(x,np.sin(x),"g--")
plt.plot(x,np.sin(x+np.pi/2),"r-")
plt.xlim(-0.1,np.pi+0.1)
plt.ylim(-1.1,1.1)
plt.title("Sample Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```
